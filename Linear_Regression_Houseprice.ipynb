```python
dfclean = dfclean.drop(columns = ['date', 'country', 'city','statezip', 'street','Outliers','bathrooms'])
dfclean.head()
```

```python
dfclean.columns
```

```python
dfclean  = pd.get_dummies(dfclean)
dfclean.head()
```

```python
x = dfclean.drop(columns = 'price')
y = dfclean['price']

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)

print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
import statsmodels.api as sm

model = sm.OLS(ytrain, xtrain).fit()
```

```python
print(model.summary())
```

```python
pred = model.predict(xtest)

data = list(zip(ytest, pred))
```

```python
ax = sns.histplot(ytest)
```

```python
ax = sns.histplot(pred)
```

```python
comptab = pd.DataFrame(data, columns=['Actual', 'Predicted'])
comptab.head()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(ytest, pred))
print(rmse)
```
